```python
df["Career_Group"] = df["Suggested Job Role"].map(career_mapping)

# Drop rows that did not map (if any)
df = df.dropna(subset=["Career_Group"])
```

```python
print(df["Career_Group"].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

# Separate X and y
X = df.drop(["Suggested Job Role", "Career_Group"], axis=1)
y = df["Career_Group"]

# Split first (prevents leakage)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
nominal_cols = [
    'certifications',
    'workshops',
    'Interested subjects',
    'interested career area ',
    'Type of company want to settle in?',
    'Interested Type of Books',
    'Management or Technical',
    'hard/smart worker'
]


preprocessor = ColumnTransformer(
    transformers=[
        ('nominal', OneHotEncoder(handle_unknown='ignore'), nominal_cols)
    ],
    remainder='passthrough'
)
```

```python
X.select_dtypes(include='object').columns
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', BalancedRandomForestClassifier(
        n_estimators=600,
        random_state=42,
        n_jobs=-1
    ))
])



model.fit(X_train, y_train)
```

```python
print(y.unique())
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Predict probabilities
probs = model.predict_proba(X_test)

# Get class labels in correct order
classes = model.classes_
```

```python
# Get indices of top 3 probabilities per row
top3_indices = np.argsort(probs, axis=1)[:, -3:]

# Convert indices to class labels
top3_labels = classes[top3_indices]
```

```python
# Convert y_test to numpy array
y_true = y_test.values

# Check if true label is in top 3 predictions
correct_top3 = [
    y_true[i] in top3_labels[i]
    for i in range(len(y_true))
]

top3_accuracy = np.mean(correct_top3)

print("Top-3 Accuracy:", round(top3_accuracy, 4))
```
